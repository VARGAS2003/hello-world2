## Introducción al mercado financiero


"El mercado financiero es donde se realizan operaciones de compra y venta de valores." (BBVA,2015)

Según la página de BBVA el mercado financiero está integrado por diferentes tipos de mercados que se clasifican según diferentes características y propósitos de cada mercado, por ejemplo inversiones a largo plazo, inversiones a corto plazo, en que se emiten nuevos valores, en donde se compran y venden los valores existentes, etc.




**Los diferentes tipos de mercados son:**


<ol>
 <li>Capitales</li>
 <li>Dinero</li>
 <li>Primario </li>
 <li>Secundario </li>
</ol>
